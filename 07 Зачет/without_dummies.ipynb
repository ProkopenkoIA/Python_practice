 В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies?

```python
import random
import pandas as pd
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
```

```python
data = pd.DataFrame({'whoAmI':lst})
```

Без get_dummies

```python
data
```

```python
df2 = pd.DataFrame({'whoAmI':['human', 'robot'],'human':['1','0'],'robot':['0','1']})

data.join(df2.set_index('whoAmI'), on ='whoAmI')
```
